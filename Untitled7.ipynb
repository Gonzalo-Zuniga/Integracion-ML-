```python
from diagrams import Diagram, Cluster
from diagrams.aws.network import VPC
from diagrams.aws.storage import S3
from diagrams.aws.analytics import Glue, Athena
from diagrams.aws.ml import Sagemaker
from diagrams.aws.integration import SNS
from diagrams.aws.general import General

# Diagrama general
with Diagram(
    "Pipeline Anomalias - AWS",
    filename="aws_ml_anomaly",
    show=False,
    direction="LR"  # izquierda -> derecha
):
    # Fuente externa (sistemas, BD on-prem, etc.)
    source = General("Sistemas\nFuente")

    with Cluster("AWS Cloud"):
        with Cluster("VPC"):
            # S3 buckets
            s3_input = S3("S3 Input\n(Datos crudos)")
            s3_output = S3("S3 Output\n(Datos procesados)")

            # Glue para ETL
            glue = Glue("AWS Glue\nETL")

            # SageMaker para ML
            sm = Sagemaker("SageMaker\nProcesamiento ML")

            # Athena para consultar resultados
            athena = Athena("Amazon Athena\nConsultas")

            # SNS para alertas
            sns = SNS("Amazon SNS\nNotificación correo")

    # Flujo de datos
    source >> s3_input     # carga de datos a S3 (input)
    s3_input >> glue       # Glue lee datos crudos
    glue >> s3_output      # Glue escribe datos procesados

    s3_output >> sm        # SageMaker usa datos procesados
    sm >> athena           # resultados disponibles para consultas en Athena

    # Las predicciones == 1 (anomalias) van a SNS (correo)
    sm >> sns
```
